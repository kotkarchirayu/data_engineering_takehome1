```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
spark = SparkSession.builder. \
    appName("pyspark-1"). \
    getOrCreate()
```

### Read data

```python
df = spark.read.csv("/dataset/nyc-jobs.csv", header=True)
# df.printSchema()
```

### 1. Data Exploration

ðŸ”¹ 1.1 Basic Statistics

```python
display(df.describe().toPandas())
```

ðŸ”¹ 1.2 Null Value Analysis

```python
total_rows = df.count()

null_df = df.select([
    (count(when(col(c).isNull(), c)) / total_rows * 100).alias(c)
    for c in df.columns
])

display(null_df.toPandas())
```

### 2. Data Cleaning

ðŸ”¹ 2.1 Standardize Column Names

```python
import re

def clean_column_names(df):
    for c in df.columns:
        new_name = re.sub(r"[ /]", "_", c.lower())
        df = df.withColumnRenamed(c, new_name)
    return df

df_clean = clean_column_names(df)
# df_clean.printSchema()
```

ðŸ”¹ 2.2 Convert Numeric Columns

```python
def convert_numeric_columns(df):
    df = df.withColumn(
        "salary_range_from",
        regexp_replace(col("salary_range_from"), "[^0-9.]", "").cast("double")
    )
    
    df = df.withColumn(
        "salary_range_to",
        regexp_replace(col("salary_range_to"), "[^0-9.]", "").cast("double")
    )
    
    df = df.withColumn(
        "number_of_positions",
        col("#_of_positions").cast("int")
    )
    
    df = df.drop("#_of_positions")
    
    return df

df_clean = convert_numeric_columns(df_clean)
# df_clean.printSchema()
```

ðŸ”¹ 2.3 Convert Date Columns

```python
def convert_dates(df):
    df = df.withColumn(
        "posting_date",
        to_timestamp(col("posting_date"))
    )
    
    df = df.withColumn(
        "posting_updated",
        to_timestamp(col("posting_updated"))
    )
    
    df = df.withColumn(
        "process_date",
        to_timestamp(col("process_date"))
    )
    
    # Convert timestamp â†’ date
    df = df.withColumn("posting_date", col("posting_date").cast("date"))
    df = df.withColumn("posting_updated", col("posting_updated").cast("date"))
    df = df.withColumn("process_date", col("process_date").cast("date"))
    
    return df
```

```python
df_clean = convert_dates(df_clean)
```

ðŸ”¹ 2.4 Create Salary Average (Feature Engineering #1)

```python
df_clean = df_clean.withColumn(
    "salary_avg",
    (col("salary_range_from") + col("salary_range_to")) / 2
)

df_clean = df_clean.filter(col("salary_avg").isNotNull())
```

ðŸ”¹ 2.5 Encode Degree Requirement (Feature Engineering #2)

```python
df_clean = df_clean.withColumn(
    "degree_score",
    when(col("minimum_qual_requirements").contains("PhD"), 5)
    .when(col("minimum_qual_requirements").contains("Master"), 4)
    .when(col("minimum_qual_requirements").contains("Bachelor"), 3)
    .otherwise(1)
)
```

ðŸ”¹ 2.6 Add Posting Age (Feature Engineering #3)

```python
df_clean = df_clean.withColumn(
    "posting_age_days",
    datediff(current_date(), col("posting_date"))
)
```

ðŸ”¹ 2.7 Remove Duplicates

```python
df_clean = df_clean.dropDuplicates()
```

ðŸ”¹ 2.8 Cache df_clean After Cleaning

```python
df_clean.cache()
```

### 3. Store Processed Data

```python
df_clean.write.mode("overwrite").parquet("/dataset/output/nyc_jobs_processed.parquet")
```

### 4. KPI Calculations

ðŸ”¹KPI 1: Top 10 Job Categories

```python
top_categories = df_clean.groupBy("job_category") \
    .count() \
    .orderBy(col("count").desc()) \
    .limit(10)

display(top_categories.toPandas())
```

```python
import matplotlib.pyplot as plt

pdf = top_categories.toPandas()

plt.figure(figsize=(10,5))
plt.bar(pdf["job_category"], pdf["count"])
plt.xticks(rotation=90)
plt.title("Top 10 Job Categories")
plt.xlabel("Job Category")
plt.ylabel("Number of Postings")
plt.show()
```

ðŸ”¹ KPI 2: Salary Distribution per Category

```python
salary_dist = df_clean.groupBy("job_category") \
    .agg(avg("salary_avg").alias("avg_salary")) \
    .orderBy(col("avg_salary").desc())

display(salary_dist.toPandas())
```

```python
# Best Chart: Horizontal Bar Chart

pdf = salary_dist.limit(10).toPandas()

plt.figure(figsize=(12,7))   # Bigger figure

plt.barh(pdf["job_category"], pdf["avg_salary"])

plt.xlabel("Average Salary")
plt.title("Average Salary per Job Category")

plt.gca().invert_yaxis()

plt.show()
```

ðŸ”¹ KPI 3: Correlation Between Degree and Salary

```python
correlation = df_clean.stat.corr("degree_score", "salary_avg")
print("Correlation between Degree Score and Salary:", correlation)
```

```python
# Scatter Plot
pdf = df_clean.select("degree_score", "salary_avg").toPandas()

plt.figure(figsize=(6,5))
plt.scatter(pdf["degree_score"], pdf["salary_avg"], alpha=0.3)
plt.xlabel("Degree Score")
plt.ylabel("Average Salary")
plt.title("Correlation Between Degree and Salary")
plt.show()
```

ðŸ”¹ KPI 4: Highest Salary per Agency

```python
highest_salary = df_clean.groupBy("agency") \
    .agg(max("salary_avg").alias("max_salary")) \
    .orderBy(col("max_salary").desc())

display(highest_salary.toPandas())
```

```python
# Horizontal Bar Chart (Top 10 Agencies)
pdf = highest_salary.limit(10).toPandas()

plt.figure(figsize=(10,6))
plt.barh(pdf["agency"], pdf["max_salary"])
plt.xlabel("Maximum Salary")
plt.title("Highest Salary per Agency")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

ðŸ”¹ KPI 5: Average Salary per Agency (Last 2 Years)

```python
from pyspark.sql.functions import max, add_months, lit, avg

# Step 1: Get max date from dataset
max_date = df_clean.select(max("posting_date")).first()[0]

# Step 2: Calculate cutoff date (2 years before max_date)
cutoff_date = spark.range(1) \
    .select(add_months(lit(max_date), -24).alias("cutoff")) \
    .first()[0]

# Step 3: Filter and compute KPI
avg_salary_last_2_years = df_clean.filter(
    col("posting_date") >= cutoff_date
).groupBy("agency") \
 .agg(avg("salary_avg").alias("avg_salary_last_2_years")) \
 .orderBy(col("avg_salary_last_2_years").desc())

display(avg_salary_last_2_years.toPandas())
```

```python
# Bar Chart
pdf = avg_salary_last_2_years.limit(10).toPandas()

plt.figure(figsize=(10,6))
plt.barh(pdf["agency"], pdf["avg_salary_last_2_years"])
plt.xlabel("Average Salary")
plt.title("Average Salary per Agency (Last 2 Years)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

ðŸ”¹ KPI 6: highest paid skills in the US market

<!-- Step A â€” Clean preferred_skills -->

```python
# Step A â€” Clean preferred_skills
from pyspark.sql.functions import lower, regexp_replace

df_skills = df_clean.withColumn(
    "preferred_skills",
    regexp_replace(col("preferred_skills"), "[^a-zA-Z0-9,. ]", "")
)
```

```python
# Step B â€” Split & Explode
from pyspark.sql.functions import split, explode, trim

df_skills = df_skills.withColumn(
    "skill",
    explode(split(col("preferred_skills"), "[,.]"))
)

df_skills = df_skills.withColumn("skill", trim(col("skill")))
```

```python
# Step C - Remove Small & Large sentences
from pyspark.sql.functions import length

df_skills = df_skills.filter(
    (col("skill").isNotNull()) &
    (length(col("skill")) > 3) &
    (length(col("skill")) < 60)
)
```

```python
# Step D â€” Compute Highest Paid Skills
from pyspark.sql.functions import avg, count

highest_paid_skills = df_skills.groupBy("skill") \
    .agg(
        avg("salary_avg").alias("avg_salary"),
        count("*").alias("frequency")
    ) \
    .orderBy(col("avg_salary").desc()) \
    .limit(20)

display(highest_paid_skills.toPandas())
```

```python
# Bar Chart
pdf = highest_paid_skills.toPandas()

plt.figure(figsize=(8,6))
plt.barh(pdf["skill"], pdf["avg_salary"])
plt.xlabel("Average Salary")
plt.title("Top Highest Paid Skills")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
